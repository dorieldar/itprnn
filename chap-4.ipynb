```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import comb
from sympy import *
import math
```

```python

def binomial(N,p):
    n = np.arange(0,N+1)
    occurences  = comb(N,n)
    probs = np.multiply(np.power(p,n),np.power(1-p,N-n))
    return np.multiply(probs, occurences) 
def Hdelta(N,p):
    
    P = binomial(N,p)
    occurences  = comb(N,np.arange(0,N+1))
    return 1- np.cumsum(P), np.log2(np.cumsum(occurences))
#ex 4.15 
def plotH(Ns,p):
    for N in Ns:
        delta,H = Hdelta(N,p)
        plt.plot(delta,H/N, label=N)
    plt.xlabel('delta')
    plt.ylabel('1/N*H(X)')
    plt.xlim(0.,1.)
    plt.legend()
    plt.show()

plotH([1,2,10,100,1000],0.2)

#ex 4.16
plotH([1,2,10,100,1000],0.5)
```
