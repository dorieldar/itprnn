```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom
import math
```

```python
#eq 3.8
def bentCoinSample(n,p, f):
    return np.power(p,f)*np.power(1-p, n-f)
# eq3.12
def bentCoinSum(n,f):
    return math.factorial(f)*math.factorial(n-f) / math.factorial(n+1)

def laplaceRule(n,Fa):
    return (Fa +1)/(n+2)

#ex 3.5
def bcPosterior(n,p,f):
    return bentCoinSample(n,p,f) / bentCoinSum(n,f)

p = np.linspace(0,1)
plt.plot(p, bcPosterior(n=3,f=2,p=p))
plt.show()

plt.plot(p, bcPosterior(n=3,f=0,p=p))
plt.show()
```
